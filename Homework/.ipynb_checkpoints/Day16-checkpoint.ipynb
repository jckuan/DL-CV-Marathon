## Data Augmentation

    1. Implement data augmentation with `ImageDataGenerator` and imgaug.

### (a)

```python
from keras.preprocessing.image import ImageDataGenerator
import cv2
import numpy as np
import matplotlib.pyplot as plt

train_datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
img = cv2.imread('Lenna_(test_image).png')
img = cv2.resize(img, (224,224))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_orig = img.copy()

img = np.array(img, dtype=np.float32)
img_combine = np.array([img, img, img, img, img], dtype=np.uint8)          # make input 5D: (224, 224, 3) -> (5, 224, 224, 3)
img_new = train_datagen.flow(img_combine, batch_size=5)
assert next(img_new).shape==(5, 224, 224, 3)          # next?

plt.figure(figsize=(20, 20))
i = 1
for batch in img_new:
    plt.subplot(1, 6, i)
    plt.imshow(batch[0, :, :, :].astype(np.uint8))
    plt.imshow(batch[1, :, :, :].astype(np.uint8))
    plt.imshow(batch[2, :, :, :].astype(np.uint8))
    plt.imshow(batch[3, :, :, :].astype(np.uint8))
    plt.imshow(batch[4, :, :, :].astype(np.uint8))
    plt.axis('off')
    i += 1
    if i > 5:
        break
```

```python
import imgaug as ia
from imgaug import augmenters as iaa

# same procedure as above
# img = cv2.imread('Lenna_(test_image).png')
# img = cv2.resize(img, (224,224))
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# img_orig = img.copy()

# img = np.array(img, dtype=np.float32)
img_aug = np.random.randint(0, 255, (5, 224, 224, 3), dtype=np.uint8)

flipper = iaa.Fliplr(1.0)          # flip 100& of the images
img_aug[0] = flipper.augment_image(img) 

vflipper = iaa.Flipud(0.4)
img_aug[1] = vflipper.augment_image(img) 

blurer = iaa.GaussianBlur(sigma=(0.0, 3.0))
img_aug[2] = blurer.augment_image(img)

translater = iaa.Affine(translate_px={"x": -16})          # translate by pixels
img_aug[3] = translater.augment_image(img) 

scaler = iaa.Affine(scale={"y": (0.8, 1.2)})          # scale to 80% to 120% of orig size
img_aug[4] = scaler.augment_image(img)

plt.figure(figsize=(20, 20))
i = 1
for image in img_aug:
    plt.subplot(1, 6, i+1)
    plt.imshow(image.astype(np.uint8))
    plt.axis('off')
    i += 1
```

### (b)

```python
# similar procedure as above
# img = cv2.imread('Lenna_(test_image).png')
# img = cv2.resize(img, (224,224))
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# img_orig = img.copy()

# img = np.array(img, dtype=np.float32)
# img_combine = np.array([img, img, img, img], dtype=np.float32)          # notice dtype=float32

sometimes = lambda aug: iaa.Sometimes(0.5, aug)          # applies the given augmenter in 50% of all cases
# define customized sequence of augmentation
seq = iaa.Sequential(
    [
        iaa.Crop(px=(0, 16)),
        iaa.Fliplr(0.4),
        
        sometimes(iaa.CropAndPad(percent=(-0.05, 0.1), pad_mode=ia.ALL, pad_cval=(0, 255))),
        
        # - order: use nearest neighbor or bilinear interpolation (fast)
        sometimes(iaa.Affine(
            scale={"x": (0.8, 1.2), "y": (0.8, 1.2)},
            translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)},
            rotate=(-45, 45),
            shear=(-16, 16),
            order=[0, 1],
            cval=(0, 255),
            mode=ia.ALL
        )),
        
        # convert images into their superpixel representation,
        # sample between 20 and 200 superpixels per image, but do not replace all superpixels with their average
        sometimes(iaa.Superpixels(p_replace=(0, 1.0), n_segments=(20, 200))),
        
        sometimes(iaa.OneOf([
            iaa.GaussianBlur((0, 3.0)),
            iaa.AverageBlur(k=(2, 7)),
            iaa.MedianBlur(k=(3, 11)),
        ])),
        
        sometimes(iaa.Sharpen(alpha=(0, 1.0), lightness=(0.75, 1.5))),
        
        sometimes(iaa.Emboss(alpha=(0, 1.0), strength=(0, 2.0))),
    ],
    # do the augmentations above in random order
    random_order=True
)

img_custaug = seq.augment_images(img_combine)

plt.figure(figsize=(20, 20))
i = 1
for image in img_custaug:
    plt.subplot(1, 6, i+1)
    plt.imshow(image.astype(np.uint8))
    plt.axis('off')
    i += 1
```
